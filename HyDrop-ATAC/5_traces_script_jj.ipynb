```python
def plot_genome_region(
    bigwigs,
    selected_regions,
    gtf_file="/user/leuven/316/vsc31650/jjans/resources/FlyBase/FB2020_02/dmel-all-r6.33_chr_name.gtf",
    subtracks=None,
    colors=None,
    ymin=0,
    prefix="",
    figsize=(10, 10),
    N=20,
    saveas=None,
    subsample=None,
    ymax_pergene={},
):
    # load libraries and functions
    import matplotlib as mpl
    import matplotlib.gridspec as gridspec
    import matplotlib.patheffects as PathEffects
    import matplotlib.pyplot as plt
    import numpy as np
    import pandas as pd
    import pyBigWig
    import seaborn as sns
    from adjustText import adjust_text
    from matplotlib.font_manager import FontProperties
    from pybedtools import BedTool

    # Smooth plot
    def running_mean(x, N):
        cumsum = numpy.cumsum(numpy.insert(x, 0, 0))
        return (cumsum[N:] - cumsum[:-N]) / float(N)

    # Round up to nearest multiple of base
    def myround(x, base=50):
        return base * round((x + base / 2 - 1) / base)

    # Take every nth element (for subsampling)
    def altElement(a, n):
        return a[::n]

    import re

    markeredge_color = "k"
    markeredge_width = 0.2
    # get values from regions
    regions_to_plot = list(selected_regions.values())
    genes_to_plot = list(selected_regions.keys())
    celltypes = [re.sub(prefix, "", x) for x in bigwigs]
    # print(celltypes)
    # convert to bed format
    index_bedtool = []
    for region, gene in zip(regions_to_plot, genes_to_plot):
        chrom, start, end = re.split(":|-", region)
        start, end = np.int(re.sub(",", "", start)), np.int(re.sub(",", "", end))
        index_bedtool.append(
            dict(chrom=chrom, start=start, end=end, name=str(region), gene=gene)
        )
    index_bedtool = pd.DataFrame(index_bedtool)
    b = BedTool.from_dataframe(df=index_bedtool)

    
    
    # calculate ymax per col
    ymax_dict = {}
    for celltype, file in zip(celltypes, bigwigs):
        ymax_dict_subcell = {}
        # load bedgraph file
        # print(file)
        bw = pyBigWig.open(file)
        # loop over regions/genes and plot them
        for region, gene in zip(regions_to_plot, genes_to_plot):
            chrom, start, end = re.split(":|-", region)
            start, end = np.int(re.sub(",", "", start)), np.int(re.sub(",", "", end))
            # chrom = re.sub("chr","",chrom) # this line makes sure that chrom matches the bigwig file!
            y = bw.values(chrom, start, end)
            # add smoothing
            # if len(y) > 100:
            #     y_convolved = np.convolve(y, np.ones((N,)) / N, mode="valid")
                # y_convolved = running_mean(y,N) #faster than convolution since convolution uses FFT
            #else:
            y_convolved = y

            y_convolved = np.array(y_convolved)
            y_convolved = np.nan_to_num(y)
            ymax_dict_subcell[gene] = y_convolved.max()
            # print(celltype)
            # print(gene)
            # print(max(y_convolved))
        ymax_dict[celltype] = ymax_dict_subcell
    j = 1

    ymax_pergene = {}
    for gene in enh_dict.keys():
        test = np.array([ymax_dict[key][gene] for key in ymax_dict.keys()])
        test = np.nan_to_num(test)
        ymax_pergene[gene] = max(test)
    # print(ymax_dict)
    # print(ymax_pergene)


    # load subtracks
    subtracks_to_plot = {}
    i = 0
    if subtracks is not None:
        for track in subtracks:
            c = BedTool(track)
            b_and_c = c.intersect(b, wa=True, wb=True)
            b_and_c = b_and_c.to_dataframe()
            b_and_c = b_and_c.T.reset_index().T
            subtracks_to_plot[i] = b_and_c
            i += 1
    nsubtracks = len(subtracks_to_plot.keys())
    # load gtf file
    c = BedTool(gtf_file)
    b_and_c = c.intersect(b, wa=True, wb=True)
    b_and_c = b_and_c.to_dataframe()
    gtf_track = b_and_c.T.reset_index().T
    # make figure
    fig = plt.figure(figsize=figsize)
    # make grids:
    # make main grid
    bottom_row = (len(celltypes) - 1) * len(regions_to_plot)
    inner_nrow, inner_ncol = 1 + 1 + nsubtracks, 1
    height_ratios = [len(celltypes)]
    for track in range(0, nsubtracks):
        height_ratios.append(0.3)
    height_ratios.append(0.2)  # for gtf
    inner = gridspec.GridSpec(
        inner_nrow, inner_ncol, wspace=0.25, hspace=0.17, height_ratios=height_ratios
    )
    # make grid for celltypes x regions
    outer_nrow, outer_ncol = len(celltypes), len(regions_to_plot)
    outer = gridspec.GridSpecFromSubplotSpec(
        outer_nrow, outer_ncol, wspace=0.05, hspace=0.2, subplot_spec=inner[0]
    )
    # initalize counter
    l = 0  # count element in celltype x region grid
    color_iter = 0
    for celltype, file in zip(celltypes, bigwigs):
        # get colors
        if colors is not None:
            color = colors[color_iter]
            color_iter += 1
        else:
            color = colors
        # load bedgraph file
        # print(file)
        bw = pyBigWig.open(file)
        # loop over regions/genes and plot them
        for region, gene in zip(regions_to_plot, genes_to_plot):
            chrom, start, end = re.split(":|-", region)
            start, end = np.int(re.sub(",", "", start)), np.int(re.sub(",", "", end))
            # chrom = re.sub("chr","",chrom) # this line makes sure that chrom matches the bigwig file!
            y = bw.values(chrom, start, end)
            x = np.array(range(start, end, 1))
            xmin, xmax = x.min(), x.max()
            if subsample:
                y = altElement(y, subsample)
                x = altElement(x, subsample)
            # add smoothing
            if len(y) > 100:
                y_convolved = np.convolve(y, np.ones((N,)) / N, mode="valid")
                # y_convolved = running_mean(y,N) #faster than convolution since convolution uses FFT
                x_convolved = x[N - 1 :]
            else:
                x_convolved = x
                y_convolved = y
            # add subplot for region for that cell type (counter = l)
            ax1 = plt.Subplot(fig, outer[l])
            if color:
                ax1.fill_between(
                    x_convolved,
                    y1=y_convolved,
                    y2=0,
                    step="mid",
                    color=color,
                    linewidth=0.5,
                    edgecolor="k",
                )
            else:
                ax1.fill_between(
                    x_convolved,
                    y1=y_convolved,
                    y2=0,
                    step="mid",
                    linewidth=0.5,
                    edgecolor="k",
                )
            ax1.set_ylim([ymin, ymax_pergene[gene]])
            ax1.set_xlim([xmin, xmax])
            ax1.set_yticks([ymin, ymax_pergene[gene]])
            left_despine = False
            if l % len(regions_to_plot) != 0:
                ax1.set_yticks([])
                left_despine = True
            if l % len(regions_to_plot) == len(regions_to_plot) - 1:
                ax1.yaxis.set_label_position("right")
                ax1.set_ylabel(celltype, rotation=0, horizontalalignment="left")
            if l < bottom_row:
                ax1.set_xticks([])
            ax1.ticklabel_format(style="sci", axis="x", scilimits=(0, 0))
            ax1.set_xticks([])
            ax1.set_yticks([])
            sns.despine(top=True, right=True, left=left_despine, ax=ax1)
            fig.add_subplot(ax1)
            l += 1

    j = 1

    gtf_grid = gridspec.GridSpecFromSubplotSpec(
        1, outer_ncol, wspace=0.05, hspace=0, subplot_spec=inner[j]
    )
    b_and_c = gtf_track
    # loop over regions
    z = 0
    for gene in genes_to_plot:
        print(gene)
        b_and_c_subset = b_and_c.loc[b_and_c[13] == gene]  # subset to region/gene
        b_and_c_subset_gene = b_and_c_subset.loc[
            b_and_c_subset[1] == "refGene"
        ]  # only keep gtf entries for gene
        ax2 = plt.Subplot(fig, gtf_grid[z])
        if len(b_and_c_subset_gene) > 0:
            gene_orientation = b_and_c_subset_gene.head(n=1)[6].values[
                0
            ]  # get orientation of gene: + or -

            chrom, start, end = re.split(":|-", region)
            start, end = np.int(re.sub(",", "", start)), np.int(re.sub(",", "", end))
            # chrom = re.sub("chr","",chrom)
            y = bw.values(chrom, start, end)
            x = np.array(range(start, end, 1))
            xmin = start
            xmax = end

            x_gene_TSS, x_gene_END = (
                b_and_c_subset_gene[10].iloc[0],
                b_and_c_subset_gene[11].iloc[0],
            )  # get TSS and END
            ax2.plot([x_gene_TSS, x_gene_END], [1, 1], linewidth=1, c="k")
            # print(gene_orientation)
            if "exon" in list(b_and_c_subset[2]):
                b_and_c_subset_exons = b_and_c_subset.loc[b_and_c_subset[2] == "exon"]
                for i in b_and_c_subset_exons.index:
                    x_exon_START, x_exon_END = (
                        b_and_c_subset_exons.loc[i, 3],
                        b_and_c_subset_exons.loc[i, 4],
                    )
                    ax2.plot([x_exon_START, x_exon_END], [1, 1], linewidth=5, c="k")
        else:
            ax2.plot(
                [0, 0], [1, 1], linewidth=0, c="k"
            )  # make empty plot (linewidth=0)

        # ax2.set_xlim([xmin,xmax])
        ax2.set_xticks([])
        ax2.set_yticks([])
        sns.despine(left=True, bottom=True, top=True, right=True, ax=ax2)
        # ax2.axes.get_xaxis().set_Tshz2isible(False)
        ax2.set_xlabel(gene, fontsize=10)
        fig.add_subplot(ax2)

        z += 1
    if saveas:
        plt.savefig(
            saveas, dpi=350, bbox_inches="tight", pad_inches=0, transparent=False
        )
        plt.show()
        plt.close()
        print(ymax_pergene)
    else:
        plt.show()
```
