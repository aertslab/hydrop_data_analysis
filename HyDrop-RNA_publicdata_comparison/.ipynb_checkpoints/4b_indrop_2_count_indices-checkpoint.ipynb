Count inDrop sample indices from fastq

```bash
bc_count_raw () {
    local fastq_R2_path="${1}";
    local fastq_R2_name="${fastq_R2_path#fastq_indrop/}";
    module load mawk;
    zcat ${fastq_R2_path} \
      | mawk -F '\t' \
        '
        {
            if (NR % 4 == 2) {
                full_barcode[substr($1, 1, 8)]++;
            }
        } END {
            for (bc in full_barcode) {
                print bc "\t" full_barcode[bc] > "bc_count_raw/bc_raw.tsv";
            }
        }'


sort -k 2,2nr bc_count_raw/bc_raw.tsv > bc_count_raw/bc_raw.sorted.tsv;
mv bc_count_raw/bc_raw.sorted.tsv bc_count_raw/${fastq_R2_name%.fastq.gz}.bccount.tsv;
rm bc_count_raw/bc_raw.tsv;
}
```

```bash
ls fastq_indrop/*3.fastq.gz
```

```bash
mkdir bc_count_raw;
for R3_fastq in fastq_indrop/*3.fastq.gz
do
    bc_count_raw $R3_fastq
done
```

```bash
head -n 15 bc_count_raw/*
```

We can clearly see that the top 12 here contains the vast, vast majority of reads. This corresponds to the 12 samples that is described in the manuscript. In other words, we can use these 12 sample indices as part of the barcode. The identity of each 12 indices were also verified with Kalish et al.
